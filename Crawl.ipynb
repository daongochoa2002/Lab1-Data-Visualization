```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import json
import re
import datetime
```

```python
website='https://www.worldometers.info/coronavirus/' # url for the site 
website_url=requests.get(website).text
soup = BeautifulSoup(website_url,'html.parser')
```

```python
my_table = soup.find_all('tr')
len(my_table)
```

```python
columns = []
data_0 = my_table[0].find_all('th')
for i in range(len(data_0)):
    columns.append(data_0[i].text)
```

# Get current data

```python
now = []
for i in range(1,248):
    now.append(my_table[i].find_all('td'))
```

```python
data_2 = {}
for i in range(len(columns)):
    data_2[columns[i]] = []
    for d in now:
        data_2[columns[i]].append(d[i].text)
```

```python
len(my_table[249].find_all('td'))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(data_2)
len(df['Country,Other'].unique())
```

```python
df.head(10)
```

```python
time = datetime.datetime.now() #- datetime.timedelta(2)
df['Date'] = time
df
```

```python
df.to_csv('data/Date'+str(time)+'.csv',index=False)
```
